### Formatar tabela com dados organizados e limpos

```python
# Importar bibliotecas
import pandas as pd
import numpy as np

# Ler arquivos
df_customers = pd.DataFrame(pd.read_csv('C:/Users/patri/GitHub/An치lise_Explorat칩ria_Python/Dataset/customer_data.csv'))

df_sales = pd.DataFrame(pd.read_csv('C:/Users/patri/GitHub/An치lise_Explorat칩ria_Python/Dataset/sales_data.csv'))

# Exibir primeiras linhas
display(df_sales.head())

display(df_customers.head())
```

```python
# Definir primeiras letras como mai칰sculas
df_customers.columns = df_customers.columns.str.title()
df_sales.columns = df_sales.columns.str.title()

# Conferir tipos das colunas e se h치 valores nulos
display(df_customers.info())
display(df_sales.info())

# Visualizar m칠tricas de vari치veis quantitativas
display(df_customers.describe().round(2))
display(df_sales.describe().round(2))
```

```python
# Preencher valores nulos com a mediana
df_customers['Age'] = df_customers['Age'].fillna(df_customers['Age'].median())

# Transformar coluna com datas em tipo datetime
df_sales['Invoice_Date'] = pd.to_datetime(df_sales['Invoice_Date'], format = '%d-%m-%Y')

# Remover duplicatas
df_customers.drop_duplicates(subset = ['Customer_Id'], inplace=True)
df_sales.drop_duplicates(subset = ['Invoice_No'], inplace=True)
```

```python
# Criar coluna de valor total
df_sales['Total_Value'] = df_sales['Quantity'] * df_sales['Price']

# Criar coluna de faixa et치ria
bins = [18, 29, 41, 51, 61, 100]
labels = ['18-29', '30-40', '41-50', '51-60', '61+']
df_customers['Age_Group'] = pd.cut(df_customers['Age'], bins = bins, labels = labels, include_lowest=True)

# Unir tabelas
df_full = pd.merge(df_sales, df_customers, on = 'Customer_Id', how = 'left')

# Verificar informa칞칫es e m칠tricas de valores quantitativos da nova tabela
display(df_full.info())
display(df_full.describe().round(2))

# Ver primeiras 10 linhas da tabela
df_full.head(10)
```

```python
# Criar arquivo csv com a tabela completa
df_full.to_csv('dados_processados_completos.csv', index=False)
```

### Fazer an치lises com os dados formatados

```python
# Importar bibliotecas para cria칞칚o de gr치ficos
import matplotlib.pyplot as plt
import seaborn as sns
```

### Distribui칞칚o de pre칞o dos produtos

```python
#Definir intervalos
bins = list(range(0, int(df_full['Price'].max()) + 500, 500))

# Criar vari치vel da distribui칞칚o do pre칞o em um histograma
ax = df_full['Price'].hist(bins=bins, figsize=[12, 4], color='blue', edgecolor='black')

# Personalizar gr치fico
plt.title('Distribui칞칚o de Pre칞o')
plt.xlabel('Pre칞o')
plt.ylabel('Frequ칡ncia')
plt.xticks(bins)
plt.grid(False)

# Adicionar r칩tulos de dados
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.text(p.get_x() + p.get_width() / 2, height, f'{int(height)}', 
                ha='center', va='bottom', fontsize=10,
                bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3'))

# Exibir gr치fico
plt.show()
```

### Distribui칞칚o de Vendas por Categoria de Produto

```python
# Agrupar a quantidade vendida por categoria do produto
vendas_por_produto = df_full.groupby('Category')['Quantity'].sum()

# Criar vari치vel com o gr치fico de barras
plt.figure(figsize = [10,5])
ax2 = sns.barplot(data = vendas_por_produto, orient = 'h', color = 'royalblue', edgecolor = 'black', order = df_full['Category'].value_counts().index)

#Personalizar gr치fico
plt.title('Distribui칞칚o de Vendas por Categoria de Produto')
plt.xlabel('N칰mero de Vendas')
plt.ylabel('Categoria')

# Adicionar r칩tulos de dados
for i in ax2.patches:
    ax2.text(i.get_width(), 
             i.get_y() + i.get_height() / 2, 
             f'{int(i.get_width())}', 
             ha = 'center', va = 'center', 
             bbox=dict(facecolor='white', edgecolor='black'))
    
# Exibir gr치fico
plt.show()
```

### Vendas de m칡s a m칡s em cada ano

```python
# Criar colunas Ano e M칡s
df_full['Year'] = df_full['Invoice_Date'].dt.year
df_full['Month'] = df_full['Invoice_Date'].dt.strftime("%Y-%m")

# Vari치vel com os dados de cada ano
years = df_full['Year'].unique()

# Criar dicion치rio com os dados agrupados por m칡s com a soma do valor total para cada ano
df_ano_dict = {}
for year in years:
    df_ano_dict[year] = df_full[df_full['Year']==year].groupby('Month')['Total_Value'].sum()

# Criar espa칞os para um gr치fico para cada ano
fig, axes = plt.subplots(nrows = len(years), ncols = 1, figsize = [12, 12])

# Preencher cada gr치fico com os valores de cada ano
for ax, year in zip(axes, years):
    ax.plot(df_ano_dict[year], linewidth = 3.5, color = 'cornflowerblue', marker = 'o')
    fig.suptitle("Valor Total M칡s a M칡s", fontsize=16, fontweight='bold')
    ax.set_xlabel('Month')
    ax.set_ylabel('Total Value')
    ax.legend([f"Ano {year}"])
    ax.grid(True)
    plt.tight_layout()

    # Adicionar r칩tulos de dados
    for i, value in enumerate(df_ano_dict[year]):
        ax.text(x = i, y = value, s = f'{value:,.2f}', 
                ha='center', va='bottom', fontsize=10,
                bbox = dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3'))

#Exibir gr치fico
plt.show()
```

### Quantidade de Produtos Vendidos por Faixa Et치ria e Categoria de Produto

```python
# Definir paleta de cores
sns.set_palette("deep")

# Somar a quantidade de vendas por faixa et치ria e categorias
vendas_por_idade_categoria = df_full.groupby(['Age_Group', 'Category'])['Quantity'].sum().unstack()

# Criar gr치fico de barras acumulado
ax = vendas_por_idade_categoria.plot(kind = 'bar', stacked = True, figsize = [12, 10])

# Personalizando o gr치fico
plt.title('Quantidade de Produtos Vendidos por Faixa Et치ria e Categoria de Produto', fontsize = 15)
plt.xlabel('Faixa Et치ria', fontsize = 12)
plt.ylabel('Quantidade Total Vendida', fontsize = 12)
plt.xticks(rotation = 0)
plt.legend(title = 'Categorias')
plt.tight_layout()

# Adicionar r칩tulos de dados
for i in ax.patches:
    ax.text(x = i.get_x() + i.get_width() / 2, 
            y = i.get_y() + i.get_height() / 2, 
            s = f'{ i.get_height():.0f}', 
            ha = 'center', va = 'center', fontsize = 12, fontweight='bold',
            bbox = dict(facecolor = 'white', edgecolor='black', boxstyle='round,pad=0.3'))

# Exibir gr치fico
plt.show()
```
