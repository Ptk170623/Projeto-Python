### Formatar tabela com dados organizados e limpos

```python
# Importar bibliotecas
import pandas as pd
import numpy as np

# Ler arquivos
df_customers = pd.DataFrame(pd.read_csv('C:/Users/patri/GitHub/Análise_Exploratória_Python/Dataset/customer_data.csv'))

df_sales = pd.DataFrame(pd.read_csv('C:/Users/patri/GitHub/Análise_Exploratória_Python/Dataset/sales_data.csv'))

# Exibir primeiras linhas
display(df_sales.head())

display(df_customers.head())
```

```python
# Definir primeiras letras como maiúsculas
df_customers.columns = df_customers.columns.str.title()
df_sales.columns = df_sales.columns.str.title()

# Conferir tipos das colunas e se há valores nulos
display(df_customers.info())
display(df_sales.info())

# Visualizar métricas de variáveis quantitativas
display(df_customers.describe().round(2))
display(df_sales.describe().round(2))
```

```python
# Preencher valores nulos com a mediana
df_customers['Age'] = df_customers['Age'].fillna(df_customers['Age'].median())

# Transformar coluna com datas em tipo datetime
df_sales['Invoice_Date'] = pd.to_datetime(df_sales['Invoice_Date'], format = '%d-%m-%Y')

# Remover duplicatas
df_customers.drop_duplicates(subset = ['Customer_Id'], inplace=True)
df_sales.drop_duplicates(subset = ['Invoice_No'], inplace=True)
```

```python
# Criar coluna de valor total
df_sales['Total_Value'] = df_sales['Quantity'] * df_sales['Price']

# Criar coluna de faixa etária
bins = [18, 29, 41, 51, 61, 100]
labels = ['18-29', '30-40', '41-50', '51-60', '61+']
df_customers['Age_Group'] = pd.cut(df_customers['Age'], bins = bins, labels = labels, include_lowest=True)

# Unir tabelas
df_full = pd.merge(df_sales, df_customers, on = 'Customer_Id', how = 'left')

# Verificar informações e métricas de valores quantitativos da nova tabela
display(df_full.info())
display(df_full.describe().round(2))

# Ver primeiras 10 linhas da tabela
df_full.head(10)
```

```python
# Criar arquivo csv com a tabela completa
df_full.to_csv('dados_processados_completos.csv', index=False)
```

### Fazer análises com os dados formatados

```python
# Importar bibliotecas para criação de gráficos
import matplotlib.pyplot as plt
import seaborn as sns
```

### Distribuição de preço dos produtos

```python
#Definir intervalos
bins = list(range(0, int(df_full['Price'].max()) + 500, 500))

# Criar variável da distribuição do preço em um histograma
ax = df_full['Price'].hist(bins=bins, figsize=[12, 4], color='blue', edgecolor='black')

# Personalizar gráfico
plt.title('Distribuição de Preço')
plt.xlabel('Preço')
plt.ylabel('Frequência')
plt.xticks(bins)
plt.grid(False)

# Adicionar rótulos de dados
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.text(p.get_x() + p.get_width() / 2, height, f'{int(height)}', 
                ha='center', va='bottom', fontsize=10,
                bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3'))

# Exibir gráfico
plt.show()
```

### Distribuição de Vendas por Categoria de Produto

```python
# Agrupar a quantidade vendida por categoria do produto
vendas_por_produto = df_full.groupby('Category')['Quantity'].sum()

# Criar variável com o gráfico de barras
plt.figure(figsize = [10,5])
ax2 = sns.barplot(data = vendas_por_produto, orient = 'h', color = 'royalblue', edgecolor = 'black', order = df_full['Category'].value_counts().index)

#Personalizar gráfico
plt.title('Distribuição de Vendas por Categoria de Produto')
plt.xlabel('Número de Vendas')
plt.ylabel('Categoria')

# Adicionar rótulos de dados
for i in ax2.patches:
    ax2.text(i.get_width(), 
             i.get_y() + i.get_height() / 2, 
             f'{int(i.get_width())}', 
             ha = 'center', va = 'center', 
             bbox=dict(facecolor='white', edgecolor='black'))
    
# Exibir gráfico
plt.show()
```

### Vendas de mês a mês em cada ano

```python
# Criar colunas Ano e Mês
df_full['Year'] = df_full['Invoice_Date'].dt.year
df_full['Month'] = df_full['Invoice_Date'].dt.strftime("%Y-%m")

# Variável com os dados de cada ano
years = df_full['Year'].unique()

# Criar dicionário com os dados agrupados por mês com a soma do valor total para cada ano
df_ano_dict = {}
for year in years:
    df_ano_dict[year] = df_full[df_full['Year']==year].groupby('Month')['Total_Value'].sum()

# Criar espaços para um gráfico para cada ano
fig, axes = plt.subplots(nrows = len(years), ncols = 1, figsize = [12, 12])

# Preencher cada gráfico com os valores de cada ano
for ax, year in zip(axes, years):
    ax.plot(df_ano_dict[year], linewidth = 3.5, color = 'cornflowerblue', marker = 'o')
    fig.suptitle("Valor Total Mês a Mês", fontsize=16, fontweight='bold')
    ax.set_xlabel('Month')
    ax.set_ylabel('Total Value')
    ax.legend([f"Ano {year}"])
    ax.grid(True)
    plt.tight_layout()

    # Adicionar rótulos de dados
    for i, value in enumerate(df_ano_dict[year]):
        ax.text(x = i, y = value, s = f'{value:,.2f}', 
                ha='center', va='bottom', fontsize=10,
                bbox = dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3'))

#Exibir gráfico
plt.show()
```

### Quantidade de Vendas por Localização

```python
# Criar figura para inserir o gráfico
plt.figure(figsize = [14, 6])

# Criar gráfico 
ax = sns.countplot(x = df_full['Shopping_Mall'], color = 'tomato', edgecolor = 'black')

# Personalizar gráfico
plt.title('Distribuição de Vendas por Localização (Shopping Mall)', fontsize = 15)
plt.xlabel('Shopping Mall', fontsize = 12)
plt.ylabel('Número de Vendas', fontsize = 12)
plt.xticks(rotation=15)

# Adicionar rótulos de dados
for i in ax.patches:
    ax.text(x = i.get_x() + i.get_width() / 2, 
            y = i.get_height(), s = f'{i.get_height():.0f}', 
            ha = 'center', va = 'bottom', fontsize = 12,
            bbox = dict(facecolor = 'white', edgecolor='black', boxstyle='round,pad=0.3'))

# Exibir gráfico
plt.show()
```

### Quantidade de Produtos Vendidos por Faixa Etária e Categoria de Produto

```python
# Definir paleta de cores
sns.set_palette("deep")

# Somar a quantidade de vendas por faixa etária e categorias
vendas_por_idade_categoria = df_full.groupby(['Age_Group', 'Category'])['Quantity'].sum().unstack()

# Criar gráfico de barras acumulado
ax = vendas_por_idade_categoria.plot(kind = 'bar', stacked = True, figsize = [12, 10])

# Personalizando o gráfico
plt.title('Quantidade de Produtos Vendidos por Faixa Etária e Categoria de Produto', fontsize = 15)
plt.xlabel('Faixa Etária', fontsize = 12)
plt.ylabel('Quantidade Total Vendida', fontsize = 12)
plt.xticks(rotation = 0)
plt.legend(title = 'Categorias')
plt.tight_layout()

# Adicionar rótulos de dados
for i in ax.patches:
    ax.text(x = i.get_x() + i.get_width() / 2, 
            y = i.get_y() + i.get_height() / 2, 
            s = f'{ i.get_height():.0f}', 
            ha = 'center', va = 'center', fontsize = 12, fontweight='bold',
            bbox = dict(facecolor = 'white', edgecolor='black', boxstyle='round,pad=0.3'))

# Exibir gráfico
plt.show()
```
